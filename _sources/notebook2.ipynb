## Equal Width Binning
``` Pendekatan binning yang paling sederhana yaitu membagi rentang X menjadi k interval yang sama lebar. Lebar interval yaitu rentang X dibagi dengan k ```

> Formula
<p align="justify">
$$
\begin{align*}
\displaystyle \text{w} &= \frac{x_{max}-x_{min}}{k}
\end{align*}
$$

````python
ir['S-sepal_length'] = pd.cut(ir['sepal_length'], 3, labels=labels)
ir['S-sepal_width'] = pd.cut(ir['sepal_width'], 3, labels=labels)
ir['S-petal_length'] = pd.cut(ir['petal_length'], 3, labels=labels)
ir['S-petal_width'] = pd.cut(ir['petal_width'], 3, labels=labels)
ir
````

## **Membuat fungsi**

#### Fungsi untuk menghitung jumlah data berdasarkan kategori

````python
def overallcat(eg, labels, col, cat):
  group = eg.groupby(cat).count()
  return [group.loc[label, col] for label in labels]
````

#### Fungsi untuk proses info dari D

````python
def split(value, eg, labels, col, cat):
  less_group = eg[eg[col] < value]
  greater_group = eg[eg[col] >= value]

  length_less_group = overallcat(less_group, labels, col, cat)
  length_greater_group = overallcat(greater_group, labels, col, cat)

  return (length_less_group, length_greater_group)
````

## Entropy
<p align="justify">

$$
\begin{align*}
\displaystyle Entropy(S) &= \sum_{i=0}^{k} -phi \ log_{2} \ phi \\
Keterangan &: \\
S &= Himpunan \ dataset  \\
k &= Banyaknya \ partisi \ S \\
phi &= Proporsi \ S_{i} \ terhadap \ S
\end{align*}
$$


````python
def entropy(d):
  return -(sum([i/sum(d) * log2(i/sum(d)) if i/sum(d) != 0 else 0 for i in d]))
````

## Gain
<p align="justify">
$$
\begin{align*}
\displaystyle 𝐺𝑎𝑖𝑛(𝐴) &= 𝐸𝑛𝑡𝑟𝑜𝑝𝑖(𝑆) − \sum_{i=1}^{k}\frac{\left| S \right|}{\left| S_{i} \right|} * 𝐸𝑛𝑡𝑟𝑜𝑝𝑖(𝑆_{i}) \\
Keterangan &: \\
𝑆 &: himpunan \ kasus \\
𝐴 &: atribut \\
k &: jumlah \ partisi \ atribut \ A \\
|𝑆𝑖| &: jumlah \ kasus \ pada \ partisi \ ke-i \\
|𝑆| &: jumlah \ kasus \ dalam \ S 
\end{align*}
$$



````python
def gain(inisial, new):
  return inisial - new
````
