# Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/hurulee/scorecard/master/data/germancredit.csv")
data.head(2)
```

```python
data_without_column_for_convert = pd.DataFrame(data, columns = ['duration.in.month','credit.amount','age.in.years'])
data_without_column_for_convert.head()
```

```python
data_ratarata=pd.get_dummies(data['status.of.existing.checking.account'])
data_ratarata.head()
```

```python
data_kredi=pd.get_dummies(data['creditability'])
data_kredi
```

```python
data_new = pd.concat([data_without_column_for_convert, data_kredi,data_ratarata], axis=1)
data_new.head()
```

```python
data_risk_rating = pd.DataFrame(data, columns = ['number.of.people.being.liable.to.provide.maintenance.for'])
data_risk_rating.head()
```
