# Implementasi K-NN

Algoritma KNN, bekerja berdasarkan jarak terpendek dari query instance ke sampel data latih untuk menentukan KNN. <br> <br>
Metode pencarian jarak yang digunakan adalah **Euclidean Distance** yaitu perhitungan jarak terdekat.
> Formula
<br> 
$$
\begin{aligned}
d(\mathbf{p}, \mathbf{q})=d(\mathbf{q}, \mathbf{p}) &=\sqrt{\left(q_1-p_1\right)^2+\left(q_2-p_2\right)^2+\cdots+\left(q_n-p_n\right)^2} \\
&=\sqrt{\sum_{i=1}^n\left(q_i-p_i\right)^2}
\end{aligned}
$$
<br>

Dengan:<br>
$d_{(i, f)}:$ Jarak dokumen ke-i ke dokumen ke-j <br>
$x_{i(n)}$ : Kata ke $n$ di dokumen ke- $i$ <br>
$x_{j(n)}:$ Kata ke $n$ di dokumen ke- $j$ <br>

```python
import pandas as pd
```

```python
url="https://raw.githubusercontent.com/diahkamalia/DataMining1/main/IRIS.csv"
df = pd.read_csv(url)
df
```

```python
X = df.iloc[:, :-1]
y = df.loc[:, "species"]
y = df['species'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5) 
knn.fit(X_train,y_train)
```

```python
import warnings
warnings.filterwarnings('ignore', category=UserWarning, append=True)
```

```python
classes={'Iris-setosa':'Setosa','Iris-versicolor':'Versicolor','Iris-virginica':'Virginica'}
X_new = [[6.0,	3.7, 6.2,	2.3]]
y_predict = knn.predict(X_new)
print(classes[y_predict[0]])
```
