>**Membuat statistik diskriptif dari data (setiap attribut yang bertipe numerik)**

```python
#information from column data numeric
c.describe()
```

```python
#median
mid = {'xmin':c['xmin'].median(),
		'xmax': c['xmax'].median(),'ymin':c['ymin'].median(),
    'ymax':c['ymax'].median()
    }

df = pd.DataFrame(mid, index=['median'])
df
```

```python
c["xmin"].median()
```

```python
c["ymax"].median()
```

```python
#mean
flt = {'xmin':c['xmin'].mean(),
		'xmax': c['xmax'].mean(),'ymin':c['ymin'].mean(),
    'ymax':c['ymax'].mean()
    }

df = pd.DataFrame(flt, index=['mean'])
df
```

```python
c["ymax"].mean()
```

```python
#modulus
mdl = {'xmin':c['xmin'].mode(),
		'xmax': c['xmax'].mode(),'ymin':c['ymin'].mode(),
    'ymax':c['ymax'].mode()
    }

df = pd.DataFrame(mdl, index=['mode'])
df
```

>**Mendeskripsikan kemiringan/skewned dari data setiap variabel (bertipe numerik)**

##Rumus Skewned
$$ \operatorname{Kurt}[X]=\mathrm{E}\left[\left(\frac{X-\mu}{\sigma}\right)^4\right]=\frac{\mathrm{E}\left[(X-\mu)^4\right]}{\left(\mathrm{E}\left[(X-\mu)^2\right]\right)^2}=\frac{\mu_4}{\sigma^4}$$

```python
sns.distplot(c['xmin'],
             kde=True,
             bins= int(100/5),
             color='#E6B36F',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':2}
             )
plt.title("Data Xmin")
  
```

```python
sns.distplot(c['ymax'],kde=True,
             bins= int(100/5),color='#E6B36F',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':2})
  
```

```python
sns.distplot(c['xmax'],kde=True,
             bins= int(100/5),color='#E6B36F',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':2})
  
```
